## Question 1:

### What is the biggest factor relating to obesity?

```python
obesity = pd.read_csv("/workspaces/DS-3021/data/ObesityDataSet-1.csv")
print(obesity.info())
```

### Pseudocode:
- rename and reorganize NObeyesdad 
- group by obesity levels
- calculate average weight per obesity level
- Split Gender column into total number men and women per category
- count percent w family history per category
- count percent smokers per category

### Code

```python
obesity = obesity.drop(columns=['NCP'])
obesity.head()
```

```python
#change obesity to categorical
obesity['NObeyesdad'] = obesity['NObeyesdad'].astype("category")
obesity = obesity.rename(columns={'NObeyesdad': 'Obesity_Level'})
obesity.head()
```

```python
#change yes/no variables to 1s and 0s
obesity['family_history_with_overweight'] = obesity['family_history_with_overweight'].replace({'yes':1,'no':0})

obesity['SMOKE'] = obesity['SMOKE'].replace({'yes':1,'no':0})

#Change gender to 1 = man 0 = woman
obesity['Gender'] = obesity['Gender'].replace({'Male':1,'Female':0})

obesity.head()
```

```python
obesity1 = obesity.groupby(['Obesity_Level'])

obesity1.head()
```

```python
#make new dataframe

obesitydf = pd.DataFrame()

obesitydf = obesitydf.assign(male=obesity1.Gender.sum(),
                             avg_age=obesity1.Age.mean(),
                             avg_height=obesity1.Height.mean(),
                             smoker=obesity1.SMOKE.sum(),
                             total=obesity1.Obesity_Level.count(),
                             obesity_in_family=obesity1.family_history_with_overweight.sum())

obesitydf.head()
```

```python
#calculate percentage of male, smoker, and obesity in family

obesitydf = obesitydf.assign(male_perc = lambda x: (x['male']/x['total']),
                             smoker_perc = lambda x: (x['smoker']/x['total']),
                             family_history = lambda x: (x['obesity_in_family']/x['total']))

obesitydf.head()
```

```python
#delete previous columns
obesitydf = obesitydf.drop(columns=['male','smoker','obesity_in_family'])

obesitydf.head()
```

### Function

```python
#make function

def obesityfactors(x):
    obesity1 = x.groupby(['Obesity_Level'])
    factor = pd.DataFrame()
    factor = factor.assign(male=obesity1.Gender.sum(),
                             avg_age=obesity1.Age.mean(),
                             avg_height=obesity1.Height.mean(),
                             smoker=obesity1.SMOKE.sum(),
                             total=obesity1.Obesity_Level.count(),
                             obesity_in_family=obesity1.family_history_with_overweight.sum())
    factor = factor.assign(male_perc = lambda x: (x['male']/x['total']),
                             smoker_perc = lambda x: (x['smoker']/x['total']),
                             family_history = lambda x: (x['obesity_in_family']/x['total']))
    factor = factor.drop(columns=['male','smoker','obesity_in_family'])

    
    return factor

print(obesityfactors(obesity))
```

From the above dataframe, it seems that with increased obesity levels, there is not one exact factor that also increases.
